```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("C:\\Users\\pawan\\Downloads\\kidney_disease.csv", header=0, na_values="?")
```

```python
dataset.replace("?", np.NaN)
dataset.head(1)
```

```python
cleanup = {"Rbc":     {"normal": 1, "abnormal": 0},
           "Pc": {"normal": 1, "abnormal": 0},
           "Pcc": {"present": 1, "notpresent": 0},
           "Ba": {"present": 1, "notpresent": 0},
           "Htn": {"yes": 1, "no": 0},
           "Dm": {"yes": 1, "no": 0},
           "Cad": {"yes": 1, "no": 0},
           "Appet": {"good": 1, "poor": 0},
           "pe": {"yes": 1, "no": 0},
           "Ane": {"yes": 1, "no": 0}}
```

```python
dataset.replace(cleanup, inplace=True)
```

```python
dataset.fillna(round(dataset.mean(),2), inplace=True)
```
